```python
import cv2
import os
import random
import matplotlib.pyplot as plt

# Set the dataset path
dataset_path = "/content/India-Traffic-Sign-6/train"  # Change to 'valid' or 'test' as needed
image_folder = os.path.join(dataset_path, 'images')  # Folder containing image files

# List all image files in the 'images' folder
img_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]

# Function to display random images from the dataset
def visualize_samples(image_folder, img_files, num_samples=5):
    plt.figure(figsize=(12, 6))

    for i in range(num_samples):
        # Randomly pick an image file
        img_path = os.path.join(image_folder, random.choice(img_files))
        print("Image Path:", img_path)  # Debugging: print the image path

        image = cv2.imread(img_path)  # Read the image

        if image is None:
            print(f"Failed to load image: {img_path}")
            continue

        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for display
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(image)
        plt.axis('off')

    plt.show()

# Visualize sample images from the 'train' set
visualize_samples(image_folder, img_files)
```

##4. Data Pre-Processing

```python
import cv2
import os
from tqdm import tqdm

# Set the dataset path
image_folder = "/content/India-Traffic-Sign-6/train/images"
label_folder = "/content/India-Traffic-Sign-6/train/labels"
output_folder = "/content/India-Traffic-Sign-6/train/processed_images"

# Make output folder for processed images
os.makedirs(output_folder, exist_ok=True)

# Resize images and check if labels exist
def resize_images(image_folder, label_folder, output_folder, target_size=(416, 416)):
    img_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]

    for img_file in tqdm(img_files, desc="Resizing images"):
        img_path = os.path.join(image_folder, img_file)
        label_path = os.path.join(label_folder, img_file.replace(".jpg", ".txt").replace(".jpeg", ".txt").replace(".png", ".txt"))

        if os.path.exists(label_path):  # Ensure that the label file exists
            # Read the image
            image = cv2.imread(img_path)
            # Resize the image
            image_resized = cv2.resize(image, target_size)
            output_image_path = os.path.join(output_folder, img_file)
            cv2.imwrite(output_image_path, image_resized)

            # Optionally, resize the labels if needed (YOLO format stays the same)
            # In YOLO, the bounding box coordinates are relative, so resizing doesn't require changing annotations

        else:
            print(f"Label file missing for image: {img_file}")

resize_images(image_folder, label_folder, output_folder)
```

##5. Data Augemntation

```python
import os
import cv2
import albumentations as A

# Define the dataset path (adjust as needed)
dataset_path = '/content/India-Traffic-Sign-6/train/images'  # Replace with your path
output_path = '/content/augmented_images'  # Path to save augmented images

# Create the output folder if it doesn't exist
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Define the augmentation pipeline
augmentation_pipeline = A.Compose([
    A.RandomBrightnessContrast(p=0.2),
    A.HorizontalFlip(p=0.5),
    A.RandomScale(scale_limit=0.1, p=0.2),
    A.Rotate(limit=15, p=0.3),
])

# Loop through all image files in the dataset
for class_name in os.listdir(dataset_path):
    class_folder = os.path.join(dataset_path, class_name)

    # Ensure we are working with image files only (exclude non-image files)
    if os.path.isdir(class_folder):
        for img_file in os.listdir(class_folder):
            img_path = os.path.join(class_folder, img_file)

            # Check if the file is an image (adjust extensions as needed)
            if img_file.endswith(('.jpg', '.jpeg', '.png')):
                # Read the image
                image = cv2.imread(img_path)

                # Ensure the image was loaded properly
                if image is not None:
                    # Convert the image from BGR to RGB (since OpenCV uses BGR by default)
                    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

                    # Apply the augmentation pipeline to the image
                    augmented_image = augmentation_pipeline(image=image_rgb)['image']

                    # Convert back to BGR before saving
                    augmented_image_bgr = cv2.cvtColor(augmented_image, cv2.COLOR_RGB2BGR)

                    # Save the augmented image to the output folder
                    output_img_path = os.path.join(output_path, class_name, img_file)
                    os.makedirs(os.path.dirname(output_img_path), exist_ok=True)
                    cv2.imwrite(output_img_path, augmented_image_bgr)

print("Augmentation applied and images saved successfully!")
```

##7. Introducing Channel Attention to YoloV5 Backbone

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ChannelAttention(nn.Module):
    def __init__(self, channel, reduction=16):
        super(ChannelAttention, self).__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(channel, channel // reduction, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(channel // reduction, channel, bias=False),
            nn.Sigmoid()
        )

    def forward(self, x):
        b, c, _, _ = x.size()
        y = self.avg_pool(x).view(b, c)
        y = self.fc(y).view(b, c, 1, 1)
        return x * y.expand_as(x)
```
